Yearly Rank Change

```python
years = [2022, 2021, 2020, 2019]
for year in years:
    prev_year = year - 1
    df[f'Careers service rank {year}-{prev_year}'] = df[f'Careers service rank {year}'] - df[f'Careers service rank {prev_year}']
```

```python
years = [2022, 2021, 2020, 2019]
for year in years:
    prev_year = year - 1
    df[f'International course experience rank {year}-{prev_year}'] = df[f'International course experience rank {year}'] - df[f'International course experience rank {prev_year}']
```

```python
years = [2022, 2021, 2020, 2019]
for year in years:
    prev_year = year - 1
    df[f'International work mobility rank {year}-{prev_year}'] = df[f'International work mobility rank {year}'] - df[f'International work mobility rank {prev_year}']
```

```python
years = [2022, 2021, 2020, 2019]
for year in years:
    prev_year = year - 1
    df[f'Career progress rank {year}-{prev_year}'] = df[f'Career progress rank {year}'] - df[f'Career progress rank {prev_year}']
```

```python
years = [2022, 2021, 2020, 2019]
for year in years:
    prev_year = year - 1
    df[f'Value for money rank {year}-{prev_year}'] = df[f'Value for money rank {year}'] - df[f'Value for money rank {prev_year}']
```

Ranking Stability (Coefficient of Variation)

```python
rank_columns = [f'Careers service rank {year}' for year in years]
df['Careers service Rank Stability'] = df[rank_columns].std(axis=1) / df[rank_columns].mean(axis=1)
```

```python
rank_columns = [f'International course experience rank {year}' for year in years]
df['International course experience Rank Stability'] = df[rank_columns].std(axis=1) / df[rank_columns].mean(axis=1)
rank_columns = [f'International work mobility rank {year}' for year in years]
df['International work mobility Rank Stability'] = df[rank_columns].std(axis=1) / df[rank_columns].mean(axis=1)
rank_columns = [f'Career progress rank {year}' for year in years]
df['Career progress Rank Stability'] = df[rank_columns].std(axis=1) / df[rank_columns].mean(axis=1)
rank_columns = [f'Value for money rank {year}' for year in years]
df['Value for money Rank Stability'] = df[rank_columns].std(axis=1) / df[rank_columns].mean(axis=1)
```

```python
#Experience Change
```

```python
faculty_columns = [f'Faculty with doctorates (%) {year}' for year in years]
df['Faculty Change'] = df[faculty_columns].diff(axis=1).mean(axis=1)
```

```python
international_board_columns = [f'International board (%) {year}' for year in years]
df['International Board Change'] = df[international_board_columns].diff(axis=1).mean(axis=1)
women_on_board_columns = [f'Women on board (%) {year}' for year in years]
df['Women on board Change'] = df[women_on_board_columns].diff(axis=1).mean(axis=1)
female_student_columns = [f'Female students (%) {year}' for year in years]
df['Female Student Change'] = df[female_student_columns].diff(axis=1).mean(axis=1)
Female_faculty_columns = [f'Female faculty (%) {year}' for year in years]
df['Female faculty Change'] = df[Female_faculty_columns].diff(axis=1).mean(axis=1)
International_faculty_columns = [f'International faculty (%) {year}' for year in years]
df['International faculty Change'] = df[International_faculty_columns].diff(axis=1).mean(axis=1)
International_students_columns = [f'International students (%) {year}' for year in years]
df['International students Change'] = df[International_students_columns].diff(axis=1).mean(axis=1)
```

```python
# Gender Diversity Trends
gender_columns = [f'{gender} (%) {year}' for gender in ['Female students', 'Female faculty', 'Women on board'] for year in years]
for gender_column in gender_columns:
    df[f'{gender_column} Change'] = df[gender_columns].diff(axis=1).mean(axis=1)
```

```python
df.head()
```

```python
# Satisfaction Trends
satisfaction_columns = [f'Overall satisfaction {year}' for year in years]
df['Satisfaction Change'] = df[satisfaction_columns].diff(axis=1).mean(axis=1)
```

```python
# Internationalization Trends
intl_columns = [f'{intl} (%) {year}' for intl in ['International students', 'International faculty', 'International board'] for year in years]
for intl_column in intl_columns:
    df[f'{intl_column} Change'] = df[intl_columns].diff(axis=1).mean(axis=1) 
```

```python
#Career Progress vs. Salary Correlation
df['Career Progress vs. Salary Correlation'] = df['Career progress rank 2022'].corr(df['Salary percentage increase 2022'])
```

```python
# Value for Money Index
df['Value for Money Index'] = df['Value for money rank 2022'] / df['Weighted salary (US$) 2022']
```

```python
#Program Characteristics (Count)
program_characteristics = ['GMAT/GRE Required', 'Thesis Required', 'Co op/Internship: Required']

# Convert the 'Programmes/Streams Offered' column to a string and then count occurrences
for characteristic in program_characteristics:
    df[f'{characteristic} Count'] = df['Programmes/Streams Offered'].astype(str).apply(lambda x: x.count(characteristic))
```

```python
# Scholarship Availability (Count)
# Convert the 'Scholarships' column to strings and then count 'Yes' values
df['Scholarships Count'] = df['Scholarships'].astype(str).apply(lambda x: x.count('Yes'))
```
