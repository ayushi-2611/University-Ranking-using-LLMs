#### regression analysis

```python
import numpy as np
import statsmodels.api as sm
import matplotlib.pyplot as plt
```

```python
y = df['International course experience rank']
x = df['International board (%)']
```

Add a constant term (intercept) to the independent variable

```python
x = sm.add_constant(x)
```

Fit a linear regression model

```python
model = sm.OLS(y, x).fit()
```

```python
Print the regression summary
```

```python
print(model.summary())
```

Scatter plot of the data

```python
plt.scatter(df['International board (%)'], df['International course experience rank'], label='Data Points')
```

```python
plt.plot(df['International board (%)'], model.predict(), color='red', label='Regression Line')

plt.xlabel('International board (%)')
plt.ylabel('International course experience rank')
plt.legend()
plt.show()
```
