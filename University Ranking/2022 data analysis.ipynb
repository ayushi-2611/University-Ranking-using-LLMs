```python
import pandas as pd

df = pd.read_excel("C:/Users/gupta/Python jypter notebook/Data/2022.xlsx")

print(df.head())
```

```python
# Check for missing values
missing_values = df.isnull().sum()

# Handle missing values (e.g., drop rows with missing values)
df = df.dropna()
```

Exploratory Data Analysis

```python
# Summary statistics for numeric columns
summary_stats = df.describe()
print(summary_stats)
```

```python
import matplotlib.pyplot as plt

# Example: Histogram for a numeric column
plt.hist(df['Salary percentage increase'], bins=20, color='blue', alpha=0.7)
plt.xlabel('Salary percentage increase')
plt.ylabel('Frequency')
plt.title('Histogram of Salary percentage increase')
plt.show()
```

```python
import seaborn as sns

# Calculate correlation matrix
correlation_matrix = df.corr()

# Create a heatmap of the correlation matrix
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()
```

```python
# Example: Scatter plot for two numeric columns
plt.scatter(df['International board (%)'], df['Careers service rank'], alpha=0.5)
plt.xlabel('International board (%)')
plt.ylabel('Careers service rank')
plt.title('Scatter Plot of International board (%) vs. Careers service rank')
plt.show()
```

```python
# Group data by index and calculate the average rank
yearly_rank_avg = df.groupby(df.index)[['Rank in 2020', 'Rank in 2021']].mean()

# Plot trends over time
plt.figure(figsize=(10, 6))
plt.plot(yearly_rank_avg.index, yearly_rank_avg['Rank in 2020'], label='Rank in 2020', marker='o')
plt.plot(yearly_rank_avg.index, yearly_rank_avg['Rank in 2021'], label='Rank in 2021', marker='o')
plt.xlabel('Year')
plt.ylabel('Average Rank')
plt.title('School Ranking Trends (2020 vs. 2021)')
plt.legend()
plt.grid(True)
plt.show()
```

 Ranking Schools Based on Specific Criteria

```python
# Rank schools based on Overall Satisfaction in descending order
ranked_schools = df.sort_values(by='Overall satisfaction', ascending=False)
print(ranked_schools[['School Name', 'Overall satisfaction']])
```

Identifying Factors Affecting Salary Percentage Increase or Program Satisfaction

```python
import statsmodels.api as sm

# Define independent variables and dependent variable
X = df[['International board (%)', 'Faculty with doctorates (%)']]
y = df['Salary percentage increase']

# Add a constant for the intercept
X = sm.add_constant(X)

# Fit the regression model
model = sm.OLS(y, X).fit()

# Print the regression summary
print(model.summary())
```
